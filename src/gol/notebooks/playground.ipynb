```python
from visualize import evolve_and_visualize
from gol.gol import GameOfLife
from gol.visualize import parse_pattern

height = 100
width = 100
steps = 1000

initial_states = parse_pattern("../patterns/glider_gun.txt", height, width)

ca = GameOfLife(
    width=width,
    height=height,
    init_method="specified",
    initial_states=initial_states
)

anim = evolve_and_visualize(ca, steps)
```

```python
anim.save('../../../data/animation_gol.mp4', writer='ffmpeg', fps=30)
```

```python
from visualize import evolve_and_visualize_at_end

evolve_and_visualize_at_end(ca, steps)
```
