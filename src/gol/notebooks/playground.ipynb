```python
import numpy as np
from gol.visualize import evolve_and_visualize
from gol.gol import GameOfLife

height = 50
width = 50
generations = 200


glider_pattern = np.array([
    [0, 0, 1],
    [0, 0, 1],
    [1, 1, 1]
])
initial_states = np.zeros((height, width), dtype=int)
initial_states[20:23, 20:23] = glider_pattern

ca = GameOfLife(
    width=width,
    height=height,
    init_method="random"
)

anim = evolve_and_visualize(ca, generations)
```

```python
anim.save('../../../data/animation_gol.mp4', writer='ffmpeg', fps=30)
```

```python
from gol.visualize import evolve_and_visualize_at_end

evolve_and_visualize_at_end(ca, generations)
```
