```python
from wolfram.patterns import get_and_save_pattern_1, get_and_save_pattern_2

patterns = [
    get_and_save_pattern_1(20, 19),
    get_and_save_pattern_2(20, 19)
]
```

```python
mae_export_path = "../../../data/reports/wolfram-ga/varying_retention_rate/mae/pattern_{pattern_idx}.png"
mae_title = "Wolfram CA Fitness vs Retention Rates, Pattern {pattern_idx}, MAE"
```

```python
import numpy as np
import config
from reports.varying_retention_rate import get_fitness_by_varying_retention_rates

get_fitness_by_varying_retention_rates(
    ca_name=config.CA_NAME_WOLFRAM_CA,
    patterns=patterns,
    population_size=config.DEFAULT_POPULATION_SIZE,
    n_generations=config.DEFAULT_N_GENERATIONS,
    crossover_rate=config.DEFAULT_CROSSOVER_RATE,
    mutation_rate=config.DEFAULT_POPULATION_SIZE,
    retention_rates=np.arange(0.1, 1.0, 0.1),
    random_selection_rate=config.DEFAULT_RANDOM_SELECTION_RATE,
    steps=config.DEFAULT_WOLFRAM_NUM_STEPS,
    fitness_function_name=config.FITNESS_FUNCTION_MAE,
    repeats=3,
    title=mae_title,
    export_path=mae_export_path
)
```
