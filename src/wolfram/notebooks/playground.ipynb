```python
import numpy as np
from visualize import evolve_and_visualize, export_current_state
from wolfram.ca import WolframCA

width = 100
height = 100
steps = 99

init_states = np.zeros(100, dtype=int)
init_states[50] = 1

ca = WolframCA(
    width=width,
    height=height,
    rule_number=30,
    init_method="specified",
    initial_states=init_states
)

export_current_state(
    ca,
    "Rule 30 Initial Pattern",
    export_path="../../../data/wolfram-ga/demo/rule_30_init_pattern.png"
)

anim = evolve_and_visualize(ca, steps)
```

```python
anim.save('../../../data/animation_wolfram.mp4', writer='ffmpeg', fps=30)
```
