```python
from wolfram.visualize import evolve_and_visualize
from wolfram.ca import WolframCA

width = 100
height = 100
steps = 200

ca = WolframCA(
    width=width,
    height=height,
    rule_number=30,
    init_method="center_one"
)

anim = evolve_and_visualize(ca, steps)
```

```python
anim.save('../../../data/animation_wolfram.mp4', writer='ffmpeg', fps=30)
```

```python
from wolfram.visualize import evolve_and_visualize_at_end

evolve_and_visualize_at_end(ca, steps)
```
